```python
from jax import jit
import numpy as np

def norm(X):
    X = X - X.mean(0)
    return X/X.std(0)

norm_compiled = jit(norm)

X = jnp.array(np.random.rand(10000, 10))

Y = norm(X)
Y_compiled = norm_compiled(X)

np.allclose(Y, Y_compiled, atol=1E-6)
```

```python
%timeit norm(X).block_until_ready()
%timeit norm_compiled(X).block_until_ready()
```

```python
from flax import nnx
import jax
import jax.numpy as jnp
from jax import jit


class MyNeuralNetModel(nnx.Module):

    def __init__(self, input_size:int, hidden_size:int, output_size:int, *, rngs:nnx.Rngs):
        self.w1 = nnx.Param(rngs.params.uniform((input_size, hidden_size)))
        self.b1 = nnx.Param(jnp.zeros((hidden_size)))

        self.w2 = nnx.Param(rngs.params.uniform((hidden_size, output_size)))
        self.b2 = nnx.Param(jnp.zeros((output_size)))

    def __call__(self, x: jax.Array):
        y = x @ self.w1 + self.b1[None]
        y = y @ self.w2 + self.b2[None]
        return nnx.relu(y)
```

```python
mymodel = MyNeuralNetModel(input_size=5, hidden_size=10, output_size=2, rngs=nnx.Rngs(0))
```

```python
x = jnp.array(np.random.randn(3, 5))

y = mymodel(x)
```
